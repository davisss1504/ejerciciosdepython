**1.** Crea una función en Python que calcule el **mínimo común múltiplo**(número mas pequeño que es múltiplo) de dos
números.

```python
def mcm(x, y):
    z = max(x, y)                                          # Calculamos el mayor de esos números
    while True:
        if (z % x == 0) and (z % y == 0 ):
            return z   
        z += 1
print(mcm(8, 4))    
print(mcm(5, 23))
print(mcm(23, 34))
```

**2.** Crea una función en Python que calcule el **máximo común divisor**(número mas grande que divide dos números).

```python
def mcd(x,y):
    mcd = 1
    if x % y == 0:
        return y
    
    for k in range(int(y/2), 0, -1):
        if x % k == 0 and y % k ==0:
            mcd = k
            break
    return mcd

print(mcd(16,6))
print(mcd(31, 19))
print(mcd(4, 11))
```

**3. Basándote en datos aleatorios (los que quieras) crea un gráfico de tarta (pie chart)
con Mapplotlib**

```python
import matplotlib.pyplot as plt

medios_transporte = 'Vehiculos', 'Motocicletas', 'Bicicleta', 'Metro'
sizes = [45, 30, 15, 10]                                                         #Declaramos el tamaño de cada 'rebanada' y en sumatoria todos deben dar al 100%
explode = (0.1, 0, 0, 0)                                                         #En este punto señalamos que posicion debe 'resaltarse' y el valor, si se coloca 0, se omite

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=medios_transporte, autopct='%1.1f%%',     #Creamos el grafico, añadiendo los valores
        shadow=True, startangle=90)
ax1.axis('equal')                                                                #señalamos la forma, en este caso 'equal' es para dar forma circular
plt.title("Principal Medio de Transporte")
plt.legend()
plt.savefig('grafica_pastel.png')
plt.show()                                                                       #Mostrar figura
```

**5. Explica cómo combinar 4 gráficos en una única figura (subplot).**

```python
import matplotlib.pyplot as plt
import numpy as np

grafico = plt.figure(figsize=(15,15))
grafico.tight_layout()
lista=[25, 56.45, 66, 12.5]
colores = ['black', 'red', 'magenta', 'yellow', 'green']
for i in range(1,5):
    x = np.arange(0, 8, 2)
    y = lista
    ax = plt.subplot(2,2,i)
    ax.plot(x, y, color = colores[i])
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Grafica' + str(i))
```

Para crear mas figuras en una misma ventana podemos utilizar el siguiente comando:

subplot(numRows, numCols, plotNum)

- numRows = Número de filas
- numCols = Número de columnas
- plotNum = Número de gráfica
